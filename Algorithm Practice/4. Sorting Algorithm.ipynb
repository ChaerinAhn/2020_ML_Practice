# 2. 선택정렬: 처리되지 않은 데이터 중에서 가장 작은 데이터를 선택해 맨 앞에 있는 데이터와 바꾸는 것 
    * N번 만큼 가장 작은 수를 찾아서 맨 앞으로 보내야 함. 전체 연산 횟수는 
    * N + (N-1) + (N-2) +... + 2
    * (N^2 + N -2) /2 로 표현할 수 있는데, 빅오 표기법에 따라서 O(N^2)
 

```python
# 선택정렬

array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

for i in range(len(array)): # i 는 가장 앞쪽 원소 위치 
    min_index = i # 가장 작은 원소의 인덱스
    for j in range(i+1, len(array)):
        if array[min_index] > array[j]:
            min_index = j
    array[i], array[min_index] = array[min_index], array[i]  # 스와프
    print(array)
    
print(array)
```

# 3. 삽입 정렬: 처리되지 않은 데이터를 하나씩 골라 적절한 위치에 삽입 
    * 기존 첫번째 데이터는 그대로 있고 두번째 데이터가 어떤 위치로 들어갈 지를 판단
    * 삽입 정렬의 시간 복잡도는 O(N^2), 이미 정렬되어 있는 상태에서 다시 삽입정렬을 수행하면 훨씬 적게 걸릴 수 있음 O(N) 

```python
# 삽입정렬

array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

for i in range(len(array)): 
    for j in range(i, 0, -1): # 인덱스 i 부터 1까지 1씩 감소 
        if array[j] < array[j - 1]: # 한 칸씩 왼쪽으로 이동
            array[j], array[j-1] = array[j-1], array[j]
        else: #자기보다 작은 데이터 만나면 그 위치에서 멈춤
            break
    
print(array)
```

# 4. 퀵 정렬: 기준 데이터를 설정하고 그 기준보다 큰 데이터와 작은 데이터의 문제 위치를 바꾸는 방법 

    * 가장 많이 쓰이는 표준 정렬 라이브러리 알고리즘 
    * 기본적으로 첫번째 데이터를 기준데이터(Pivot)으로 설정한다 
    * 왼쪽에서부터는 피봇값보다 큰 값을, 오른쪽에서부터는 작은 값을 고른 후에 두개의 위치를 바꿔줌 
    * 엇갈리는 값이 나타났을 때는 그 값과 피봇값을 바꿈
    * 이렇게 하면 피봇값을 기준으로 왼쪽은 더 작은값, 오른쪽은 더 큰값이 모임
    * 이 작업을 분할이라고 함
    * 그리고 각 그룹별로 또 마찬가지로 퀵정렬 작업을 수행함 
    * 데이터의 범위가 분할할수록 반으로 줄어들기 때문에 점차 
    * 너비 x 높이 = N x logN = NlogN 으로 된다. 그러나 최악의 경우 O(N^2)일수도 있음
    * ex)이미 정렬이 이루어진 상태일 때가 최악     

```python
# 퀵정렬 

array = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array, start, end):
    if start >= end: # 원소가 한개인 경우 종료
        return
    pivot = start # 피벗은 첫번째 원소의 인덱스
    left = start + 1
    right = end 
    while(left <= right):
        # 피벗보다 큰 데이터를 찾을 때까지 반복
        while(left <= end and array[left] <= array[pivot]):
            left += 1
        # 피벗보다 작은 데이터를 찾을 때까지 반복
        while(right > start and array[right] >= array[pivot]):
            right -= 1
        if(left > right): # 엇갈렸다면 작은 데이터와 피벗을 교체
            array[right], array[pivot] = array[pivot], array[right]
        else: # 엇갈리지 않았다면 작은 데이터와 큰 데이터를 교체 
            array[left], array[right] = array[right], array[left]
    #분할 이후 왼쪽 부분과 오른쪽 부분에서 각자 정렬 수행
    quick_sort(array, start, right-1)
    quick_sort(array, right + 1, end)

quick_sort(array, 0, len(array) -1)
print(array)
```

```python
# 퀵 정렬 -파이썬 장점 살린 방식 

array = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array):
    # 리스트가 하나 이하의 원소만을 담고 있다면 종료
    if len(array) <= 1:
        return array
    pivot = array[0] #피벗은 첫 번째 원소
    tail = array[1:] #피벗을 제외한 리스트 
    
    left_side = [x for x in tail if x <= pivot] # 분할된 왼쪽 부분 
    right_side = [x for x in tail if x > pivot] # 분할된 오른쪽 부분
    
    # 분할 이후 왼쪽 부분과 오른쪽 부분에서 각각 정렬 수행하고, 전체 리스트 반환
    return quick_sort(left_side) + [pivot] + quick_sort(right_side)

print(quick_sort(array))
```

# 5. 계수 정렬: 특정 조건이 부합할때만 사용할 수 있지만 매우 빠름
    * 데이터의 크기 범위가 제한되어 정수 형태로 표현할 수 있을 때 사용가능
    * 데이터의 개수가 N, 데이터(양수) 중 최댓값이 K일때 최악의 경우에도 수행시간 O(N+K)를 보장 
    * 데이터 값과 동일한 인덱스가 몇번 증가했는지를 카운트 함 
    * 그 후에 각 인덱스 개수만큼 출력해나가면 됨 
    * 시간, 공간 복잡도: O(N + K)
    * 때에 따라 비효율적이기도. 0과 999,999로 단 2개만 존재하는 경우. 
    * 동일한 값을 가지는 데이터가 여러개 등장할 때 효과적
    * ex) 성적과 같이 범위가 정해지는 경우 
    

```python
# 계수정렬

array = [7, 5, 0, 2, 3, 1, 7, 9, 0]

count = [0] * (max(array) + 1)

for i in range(len(array)):
    count[array[i]] += 1 # 각 데이터에 해당하는 인덱스의 값 증가 
    
for i in range(len(count)): # 리스트에 기록된 정렬 정보 확인
    for j in range(count[i]):
        print(i, end=' ') # 띄어쓰기를 구분으로 등장 횟수만큼 출력 
```

```python
# 두 배열 원소 교체 문제 

N, K = map(int, input().split())
A = list(map(int, input().split()))
B = list(map(int, input().split()))
# N = 5
# K = 1
# A =[1, 5, 2, 4, 2]
# B =[2, 5, 3, 1, 2]
count = 0 

for i in range(len(A)): # i 는 가장 앞쪽 원소 위치 
    min_index = i # 가장 큰 원소의 인덱스
    for j in range(i+1, len(A)):
        if A[min_index] > A[j]:
            min_index = j
    A[i], A[min_index] = A[min_index], A[i] 

for i in range(len(B)): # i 는 가장 앞쪽 원소 위치 
    max_index = i # 가장 작은 원소의 인덱스
    for j in range(i+1, len(B)):
        if B[max_index] < B[j]:
            max_index = j
    B[i], B[max_index] = B[max_index], B[i] 

print(A, B)

for i in range(N):
    if A[i] < B[i]:
        A[i], B[i] = B[i], A[i]
        count += 1 
        if count == K:
            break 
print(A, B)
print(sum(A))
```
